# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
'''
Here are the five plots used in seaborn.

* Heatmap: A heatmap is a graphical representation of data where the values are represented by colors. It is often used to visualize
the correlation between two or more variables.
* Pairplot: A pairplot is a matrix of scatter plots that shows the relationship between each pair of variables in a dataset. It is a
useful way to explore the relationships between multiple variables at once.
* Scatter plot: A scatter plot is a graph that shows the relationship between two variables. The points on the scatter plot are plotted
according to their values for the two variables.\
* Line plot: A line plot is a graph that shows the trend of a variable over time. The points on the line plot are connected by lines to
show the progression of the variable over time.
* Bar plot: A bar plot is a graph that shows the distribution of a categorical variable. The bars on the bar plot represent the
frequencies of the different categories of the variable.
'''
```

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
# Code.
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Dataset.
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# Plotting the data.
sns.lineplot(x = fmri['timepoint'], y = fmri['signal'], hue = fmri['region'], style = fmri['event'])
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
# Code.
t = sns.load_dataset('titanic')
t.head()
```

```python
# PLotting first box plot.
sns.boxplot(x = t['pclass'], y = t['age'])
plt.show()
```

```python
# PLotting second box plot.
sns.boxplot(x = t['pclass'], y = t['fare'])
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
# Code.
d = sns.load_dataset('diamonds')
d.head()
```

```python
# Plotting a histogram.
sns.histplot(data = d, x = 'price', hue = 'cut', element='step', common_norm = False)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Price Distribution by Cut')
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
# Code.
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Plotting a pairplot.
sns.pairplot(iris, hue = 'species')
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Code.
f = sns.load_dataset('flights')
f.head()
```

```python
# Plotting Heatmap.
flights_pivot = f.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='coolwarm', annot=True, fmt='d')
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count Heatmap (Flights Data)')
plt.show()
```

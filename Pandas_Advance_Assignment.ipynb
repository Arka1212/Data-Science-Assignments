# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
# Python Fucntion to re-index.
def change_index(df):
  '''
  Takes a dataframe and re-indexes the dataframe
  '''
  new_index = range(1, 2 * len(df) + 1, 2)
  new_df = df.set_index(pd.Index(new_index))
  return new_df
```

```python
# Applying Fucntion to previous dataset.
change_index(df)
```

# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
# Python Function to calculate sum of first three values.
def first_three(df):
  '''
  Takes in a dataset and outputs sum of the first three values in values column.
  '''
  if 'Values' in df.columns:
    v_col = df['Values']

  # Summing up the value.
  val_sum = v_col.iloc[:3].sum()
  print(val_sum)
```

```python
# Creating sample data.
s_data = pd.DataFrame({'Values': [10,20,30,40,50]})
s_data
```

```python
# Applying the above function.
first_three(s_data)
```

# Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
# Random Text Data.
pd.set_option("display.max_colwidth", 300)        # Showing full data.

df_3 = {'Text1': ['''HTML is used to structure and organize the content on a web page. It uses various tags to define the different
                  elements on a page, such as headings, paragraphs, and links.''',
                  '''Replit is a cloud-based development environment that allows programmers to build and share projects online.''',
                  '''Replit is free to use and supports a variety of languages, including JavaScript, Python, Java, C++, and HTML/CSS.
                  It is a popular tool for beginners and experienced programmers alike.''']
        }
txt_df = pd.DataFrame(df_3)
txt_df
```

```python
# Creating a column storing word count.
def count_words(txt):
  '''
  Takes a text data and outputs the count of the words in the text
  '''
  s = str(txt)
  l = list(s.split(" "))
  return len(l)

# Applying the function to columns.
txt_df['word_count'] = txt_df['Text1'].apply(count_words)
txt_df
```

# Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

```python
# Creating a data.
e_dict = {'email': ['sayan2323@gmail.com',
                    'farooq4223@gmail.com',
                    'debashis4422@gmail.com']}

e_df = pd.DataFrame(e_dict)
e_df
```

```python
e_df['username'] = e_df['email'].apply(lambda x : x.split('@')[0])
e_df
```

# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
# Python Fucntion to re-index.
def change_index(df):
  '''
  Takes a dataframe and re-indexes the dataframe
  '''
  new_index = range(1, 2 * len(df) + 1, 2)
  new_df = df.set_index(pd.Index(new_index))
  return new_df
```

```python
# Applying Fucntion to previous dataset.
change_index(df)
```

# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.

```python
# Python Function to calculate sum of first three values.
def first_three(df):
  '''
  Takes in a dataset and outputs sum of the first three values in values column.
  '''
  if 'Values' in df.columns:
    v_col = df['Values']

  # Summing up the value.
  val_sum = v_col.iloc[:3].sum()
  print(val_sum)
```

```python
# Creating sample data.
s_data = pd.DataFrame({'Values': [10,20,30,40,50]})
s_data
```

```python
# Applying the above function.
first_three(s_data)
```

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function tocreate a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
# Sample DataFrame with 'Sales' and 'Date' columns
import numpy as np
data = {'Sales': [100, 120, 150, 130, 140, 160, 170, 180, 190, 200],
        'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
                 '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10']}

sales_data = pd.DataFrame(data)
sales_data
```

```python
# Creating a function to calculate moving average.
def moving_avg_col(df):
  df = df.sort_values(by = 'Date')
  df['MovingAverage'] = df['Sales'].rolling(window = 7, min_periods = 1).mean()
  return df


# Applying the function to dataset.
moving_avg_col(sales_data)
```

# 4 2023-01-05 Thursday

```python
# Sample DataFrame
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df_weekdays = pd.DataFrame(data)
df_weekdays
```

# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
# Code
def select_date_range(df):
    # Convert 'Date' column to datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Define the start and end dates for the desired range
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')

    # Use boolean indexing to select rows within the date range
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

    return selected_rows

# Sample DataFrame
data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10', '2023-02-28']}
df = pd.DataFrame(data)

# Call the function to select rows within the date range
selected_df = select_date_range(df)

# Print the resulting DataFrame
selected_df
```
